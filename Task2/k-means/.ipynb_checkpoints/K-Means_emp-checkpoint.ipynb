# Задание:

Реализовать алгоритм кластеризации k-means

```python
class KMeans:
    def __init__(self, n_clusters=2, metric='euclidean', max_iter=300):
        '''
        n_clusters - число кластеров
        metric - метрика
        max_iter - максимальное число итераций
        '''
        self.n_clusters = n_clusters
        self.metric = metric
        self.max_iter = max_iter

    @staticmethod
    def distance(vector1, vector2, metric='euclidean'):
        if metric == 'manhattan':
            return np.sum(np.abs(vector1 - vector2))
        elif metric == 'chebyshev':
            return np.max(np.abs(vector1 - vector2))
        else:
            return np.sqrt(np.sum((vector1-vector2)**2))
        
        '''
        Определяем функцию расстояния
        '''
        
    
    def predict(self, X):
        '''
        Предсказываем попадание объектов из X в конкретный кластер
        '''
        X = np.array(X)
        
        clusters = np.zeros(X.shape[0]).astype(int)

        for ind, point in enumerate(X):
                dists = list(map(lambda i, j: KMeans.distance(i, j, self.metric), [point]*self.n_clusters, self.centers))
                clusters[ind] = dists.index(min(dists))
                
        return X, clusters

    def fit(self, X):  
        '''
        Шаг 1 - Инизиализируем начальные положения центров кластеров
        '''
        X = np.array(X)
        n_vectors, dim = X.shape
        
        centers = np.random.randint(0, n_vectors, self.n_clusters)
        centers = np.array([np.array(X[i]) for i in centers])
        clusters = np.zeros(n_vectors)
        '''
        Шаг 2 - Выполняем уточнение положения центров кластеров до тех пор, пока 
        не будет превышено значение max_iter или центры кластеров не будут меняться 
        '''
        for step in range(self.max_iter):            
            #print(step)
            '''
            Шаг 2.1 - Вычисляем расстояние до цетров кластеров
            '''
            for ind, point in enumerate(X):
                dists = list(map(lambda i, j: KMeans.distance(i, j, self.metric), [point]*self.n_clusters, centers))
                clusters[ind] = dists.index(min(dists))
                
            '''
            Шаг 2.2 - Для каждого объекта находим argmin от расстояний до центров
            '''
            old_centers = np.array(centers)
            for ind in range(self.n_clusters):
                centers[ind] = np.sum(X[clusters == ind], axis=0)/len(X[clusters == ind])
            if np.max(list(map(lambda i, j: KMeans.distance(i, j, self.metric), old_centers, centers))) < 0.0001:
                break
            '''
            Шаг 2.3 - Уточняеням положения центров кластеров
            '''
            
        '''
        Шаг 3 - Сохраняем положения центров кластеров
        ''' 
        self.centers = centers
        '''
        Шаг 4 - Возвращаем предсказание
        '''        
        return self.predict(X)
```

* Для задачи кластеризации цветов на изображении (пример в k-means_photo_clusterization)

```python
kmeans = KMeans(n_clusters=2, metric ='chebyshev')
data=pd.read_csv('../algorithms_compare/dataset1.csv', header=None, sep='\t')
data, clusters = kmeans.fit(data)

colors = ('red', 'green', 'blue', 'violet')
colors = ([colors[i] for i in clusters])
plt.scatter(data[:,0],data[:,1], c=colors)
```
